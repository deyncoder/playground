# First Dataset

```python
# import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib.ticker import MultipleLocator
```

```python
# set file paths

first_csv_path = r'D:\DFiles\Files\Jupyter_Directory\salary_dataset.csv'

# set column names

col_names = ['years_of_experience','salary_in_php']
```

```python
# read csv where first column is index

df_salary = pd.read_csv(first_csv_path, names = col_names, index_col=0, header=0)
df_salary
```

```python
# display column info

df_salary.info()
```

```python
# basic stats

df_salary.describe()
```

```python
# number of rows

row_count = len(df_salary)
print("row count: ", row_count)

# number of columns

column_count = len(df_salary.columns)
print("column count: ", column_count)

# show list of column names

column_names = list(df_salary.columns.values)
print("column names: ", column_names)
```

# First Plot
- 1 independent variable, 1 dependent variable

```python
years_of_experience = list(df_salary['years_of_experience'])
salary_in_php = list(df_salary['salary_in_php'])
```

```python
plt.figure(figsize = (10,4))

ax = plt.axes()
ax.xaxis.set_major_locator(MultipleLocator(1))
ax.xaxis.set_minor_locator(MultipleLocator(0.1))
ax.yaxis.set_major_locator(MultipleLocator(50000))
ax.yaxis.set_minor_locator(MultipleLocator(10000))

ax.tick_params(which='both', width=1)
ax.tick_params(which='major', length=7)
ax.tick_params(which='minor', length=4)

line1 = plt.plot(years_of_experience,salary_in_php, color='lightcoral', ms=4, marker='o', mfc='indianred')
plt.ylim((0,12))
plt.ylim((0,150000))


font_title = {'family':'sans-serif','color':'y','size':15}
font_labels = {'family':'sans-serif','color':'navy','size':12}

plt.title("Experience vs Salary", fontdict=font_title)
plt.xlabel("Experience (Years)", fontdict=font_labels)
plt.ylabel("Salary (Php)", fontdict=font_labels)

plt.show()
```

# Second Dataset

```python
# set file paths

second_csv_path = r'D:\DFiles\Files\Jupyter_Directory\sales_dataset.csv'
```

```python
# read csv 

df_sales = pd.read_csv(second_csv_path)
df_sales
```

# Second Plot
- 2 plots, same x and y xis

```python
sales_month = list(df_sales['sales_month'])
total_sales_ph = list(df_sales['total_sales_ph'])
total_sales_us = list(df_sales['total_sales_us'])
```

```python
plt.figure(figsize = (10,4))

ax = plt.axes()
ax.yaxis.set_major_locator(MultipleLocator(100))
ax.yaxis.set_minor_locator(MultipleLocator(10))

ax.tick_params(which='both', width=1)
ax.tick_params(which='major', length=7)
ax.tick_params(which='minor', length=4)

line1 = plt.plot(sales_month,total_sales_ph, color='plum', ms=4, marker='o', mfc='mediumorchid')
line2 = plt.plot(sales_month,total_sales_us, color='lightcoral', ms=4, marker='o', mfc='indianred')
plt.ylim((0,550))


font_title = {'family':'sans-serif','color':'y','size':15}
font_labels = {'family':'sans-serif','color':'navy','size':12}

plt.title("Sales per Month (2022)", fontdict=font_title)
plt.xlabel("Month", fontdict=font_labels)
plt.ylabel("Number of Sales", fontdict=font_labels)
plt.legend(['PH Sales','US Sales'])
plt.show()

# plt.title("Sales per Month (2022)", fontsize=15, color='y', family='sans-serif', weight='bold')
# plt.xlabel("Month", fontsize=12, family='sans-serif')
# plt.ylabel("Number of Sales", fontsize=12, family='sans-serif')

# https://matplotlib.org/stable/gallery/color/named_colors.html
```

```python
barWidth = 0.4
plt.figure(figsize = (11,4))


ax = plt.axes()
ax.yaxis.set_major_locator(MultipleLocator(100))
ax.yaxis.set_minor_locator(MultipleLocator(10))
plt.xticks(range(0,len(sales_month)), labels=sales_month)


ax.tick_params(which='both', width=1)
ax.tick_params(which='major', length=7)
ax.tick_params(which='minor', length=4)


# Set position of bar on X axis
br1 = np.arange(len(total_sales_ph))
br2 = [x + barWidth for x in br1]

# Make the plot
hbar1 = plt.bar(br1, total_sales_ph, color ='mediumorchid', width = barWidth, edgecolor ='grey')
hbar2 = plt.bar(br2, total_sales_us, color ='indianred', width = barWidth, edgecolor ='grey')
plt.bar_label(hbar1, fontsize=6)
plt.bar_label(hbar2, fontsize=6)

font_title = {'family':'sans-serif','color':'y','size':15}
font_labels = {'family':'sans-serif','color':'navy','size':12}

plt.title("Sales per Month (2022)", fontdict=font_title)
plt.xlabel("Month", fontdict=font_labels)
plt.ylabel("Number of Sales", fontdict=font_labels)
plt.legend(['PH Sales','US Sales'])
plt.show()
```
