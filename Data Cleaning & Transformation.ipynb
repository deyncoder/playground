# Basic Data Cleaning and Transformations

* import libraries
* read csv file
* remove unnecessary columns
* remove rows with null values
* reset index
* remove duplicates
* capitalized first letter of each word in a column
* round column value up to 2 decimal place
* convert column value type to int
* join multiple columns to a new column
* split columun value by delimiter to new columns
* sort by sepcific column
* rearrange columns



```python
# import pandas librbary
# pandas in a python library is a powerful python library used to analyze and manipulate data

import pandas as pd
```

```python
# set file path

csv_path = r'C:\Users\Admin\sample-uncleaned-data.csv'
destination_path = r'C:\Users\Admin\cleaned-data.csv'
```

```python
# read csv file

df = pd.read_csv(csv_path)
df
```

```python
# remove unnecessary columns

df = df.drop(columns=['notes','hobbies'])
df
```

```python
# remove rows with null values

df = df.dropna()
df
```

```python
# reset index 
# set drop to 'true' so that old index will not be added as a new column

df = df.reset_index(drop=True)
df
```

```python
# remove duplicates

df = df.drop_duplicates()
df
```

```python
# Capitalized first letter of each word for columns firstname, middlename, lastname, address

df['first_name']=df['first_name'].str.title()
df['middle_name']=df['middle_name'].str.title()
df['last_name']=df['last_name'].str.title()
df['address']=df['address'].str.title()
df
```

```python
# salary up to 2 decimal places only

df['salary_in_php']=df['salary_in_php'].round(decimals=2)
df
```

```python
# change contactnumber column to int

df['contact_number']=df['contact_number'].astype('int64')
df
```

```python
# tranform birthdate column to a single format mm/dd/yyyy

df['birth_date'] = pd.to_datetime(df['birth_date'], dayfirst=True).dt.strftime('%m/%d/%Y')
df
```

```python
# add full name column

df['full_name'] = df[['first_name','middle_name', 'last_name']].apply(lambda x: ','.join(x), axis=1)
df
```

```python
# show column names of dataframe

cols = list(df.columns.values)
cols
```

```python
# split value of address column by comma

new = df["address"].str.split(",", n = 1, expand = True)
df['city'] = new[0]
df['zip_code'] = new[1]
df
```

```python
# drop columns firstname, middlename,lastname,address

df = df.drop(columns=['first_name','middle_name','last_name','address'])
df
```

```python
#sort by fullname

df = df.sort_values(['full_name'])
df = df.reset_index(drop=True)
df
```

```python
# rearrange columns

df = df[['full_name','birth_date','contact_number','city','zip_code','occupation','salary_in_php']]
df
```

* create function to calculate age fiven birth date
* apply calculate_age function to birth_date column, store result to new column 'age'

```python
# import datetime and date
# datetime module supplies classes for manipulating dates and times

from datetime import datetime
from datetime import date
```

```python
# create function that calculates age given date of birth

def calculate_age(dob):
    dob = datetime.strptime(dob, "%m/%d/%Y").date()
    today = date.today()
    return today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
```

```python
# apply calculate_age function on each value in birth_date column, result stored to a new column 'age'

df['age'] = df['birth_date'].apply(calculate_age)
df
```

```python
# rearrange columns

df = df[['full_name','birth_date','age','contact_number','city','zip_code','occupation','salary_in_php','salary_in_usd']]
df
```
